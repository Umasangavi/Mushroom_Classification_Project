DATA PREPROCESSING

```python
df.columns
```

ENCODING

```python
from sklearn.preprocessing import LabelEncoder
```

```python
y=df['class'].copy()
X=df.drop('class',axis=1).copy()
X
```

```python
#encoding
mapping = list()
encoder =LabelEncoder()
for column in range(len(df.columns)):
    df[df.columns[column]]= encoder.fit_transform(df[df.columns[column]])
    mapping_dict = {index :label for index, label in enumerate(encoder.classes_)}
    mapping.append(mapping_dict)
```

FEATURE SCALING

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
X=pd.DataFrame(scaler.fit_transform(X),columns=X.columns)
```

TRAIN

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.7,shuffle=True,random_state=123)
X_train
```

```python
y_train
```

LOGISTIC REGRESSION MODEL

```python
from sklearn.linear_model import LogisticRegression
```

```python
model_1=LogisticRegression()
model_1.fit(X_train,y_train)
model_1.score(X_test,y_test)*100
```

SUPPORT VECTOR CLASSIFICATION MODEL

```python
from sklearn.svm import SVC
```

```python
model_2=SVC()
model_2.fit(X_train,y_train)
model_2.score(X_test,y_test)*100
```

```python
pred_result=model_2.predict(X_test)
print(pred_result)
```

```python
from sklearn.metrics import classification_report
```

```python
report= classification_report(y_test, pred_result,zero_division=0)
```
